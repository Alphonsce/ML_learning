## Метрики машинного обучения

- Прежде всего метрики нужны для того, чтобы сравнивать различные алогритмы между собой, а разные алогоритмы могут оптимизировать разные loss'ы, 
поэтому loss является показателем качества модели лишь в рамках алгоритмов, имеющих один и тот же loss, а если вдруг мы захотим сравнить алгоритм, который оптимизировал MSE  с алгоритмом, который оптимизировал AME???? Для этого мы как раз и используем единую метрику для оценки качества.

- loss может совпадать с метрикой, но loss надо брать такой, который хорошо дифференцируется и желательно быстро вычисляется,
то есть loss надо брать таким, чтобы он позволял двигать модель в правильном направлении, например взять accuracy или ROC-AUC в качестве
lossа не представляется возможным, так как это вообще непонятные функции 

- Но! Мы можем взять некоторую функцию, которая обучает модель, а потом смотреть на качество модели при помощи метрик, которые нам удобны, например, в задачах регрессии брать MSE в качестве loss и в качестве метрики очень удобно.

#### Метрики для классификации

accuracy не всегда хороша, существует еще куча других метрик

Threshold не всегда стоит ставить просто на 0.5, например, нам гораздо важнее определить всех больных, но если мы нечаянно определим здорового как больного от этого ничего страшного не случится, определить лучший threshold можно, построив кривую ROC

Вводится метрика AUC - площадь под кривой ROC, которая проходитися по всем threshold и смотрит True positive/ False positive

#### Метрики для регрессии

В случае регрессии самый популярный вариант лосса - MSE, который мы оптимизируем через градиентный спуск, но можно брать и другие лоссы, а
чтобы сравнивать модели с разными лоссами нам надо использовать какую-то метрику, удобно MSE использовать в качестве метрики в регрессии - типо насколько близко наши предсказания находятся к их лейблам
